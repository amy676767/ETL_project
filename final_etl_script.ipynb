```python
class Sources(Base):
    __tablename__ = 'sources'
    id = Column(Integer, primary_key=True)
    name = Column(String(255))

class Inventory(Base):
    __tablename__ = 'inventory'
    id = Column(Integer, primary_key=True)
    source_id = Column(String(255)) #, ForeignKey('sources.id')
    condition = Column(String(255))
    engine = Column(String(255))
    exterior_color = Column(String(255))
    drive_type = Column(String(255))
    fuel_type = Column(String(255))
    interior_color = Column(String(255))
    make = Column(String(255))
    mileage = Column(String(255)) #should be int
    model = Column(String(255))
    price = Column(String(255))
    transmission = Column(String(255))
    seller_name = Column(String(255))
    url = Column(String(255))  
    vin = Column(String(100))  
    year = Column(String(255)) #should be int
```
